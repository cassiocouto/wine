```python
import pandas as pd
from src.models.linear_regression_numpy import LinearRegressionNumpy
from src.data.preparation import prepare_data_for_classification_as_np_arrays, normalize_dataframe
```

```python
wine_df = pd.read_csv("../data/wine_quality.csv", sep=";")
features = wine_df.drop("quality", axis=1).columns.tolist()
wine_df_normalized = normalize_dataframe(wine_df)
wine_df_normalized["quality"] = wine_df["quality"]
X_train, X_test, y_train, y_test = prepare_data_for_classification_as_np_arrays(wine_df_normalized, "quality", 42, 0.2)
```

```python
model = LinearRegressionNumpy()
model.fit(X_train, y_train, learning_rate=0.1, n_iterations=1000)
y_pred = model.predict(X_test)
mse = model.mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
# original MSE: 0.38661680382086855
# updated MSE: 0.5517613567643879 - worse!
# more interactions: 0.38649437315669577 - sliiiiiiiiiiiightly better
# much more interactions: 0.3866168038208017 - no real change; It's as good as it'll be
```
