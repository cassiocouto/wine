```python
import pandas as pd
from src.models.decision_tree_numpy import DecisionTreeNumpy
from src.data.preparation import prepare_data_for_classification_as_np_arrays
```

```python
wine_df = pd.read_csv("../data/wine_quality.csv", sep=";")
features = wine_df.drop("quality", axis=1).columns.tolist()
X_train, X_test, y_train, y_test = prepare_data_for_classification_as_np_arrays(wine_df, "quality", 42, 0.2)
```

```python
model = DecisionTreeNumpy(max_depth=5)
model.fit(X_train, y_train)
predictions = model.predict(X_test)
accuracy = (predictions == y_test).mean()
print(f"Accuracy: {accuracy:.4f}")
print(model.print_tree())
```
